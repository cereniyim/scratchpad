```python
def MakeDatasets(in_csv):
    dataframe = pd.read_csv(in_csv, 
                            index_col="Unnamed: 0")
    
    # trigger computation
    n_samples = len(dataframe)

    # split dataset into train test feel free to adjust test percentage
    idx = np.arange(n_samples)
    test_idx = idx[:n_samples // 10]
    test = dataframe.loc[test_idx]

    train_idx = idx[n_samples // 10:]
    train = dataframe.loc[train_idx]
    
    return train, test
```
